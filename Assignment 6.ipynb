```python
import pandas as pd
import datetime as dt
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import seaborn as sns
```

```python
# reads data file
df = pd.read_csv('North_Cunningham_Creek_Conduit_Bypass_Streamflow.csv')

# sets N/A data to zero, specifically for the appearence of the plot
df_dcharge = df[['Date', 'Discharge']].fillna(0) 

# sets 'Date' column as a datetime object to be processed in the plot
df_dcharge['Date'] = pd.to_datetime(df_dcharge['Date']) 

df_dcharge
```

```python
# sets style of plot
plt.figure(figsize=(15,8))
sns.set(style='darkgrid')
sns.set_context('talk')

# plots lines and legends from dataframe
ax = sns.lineplot(x=df_dcharge['Date'],
                  y=df_dcharge['Discharge'],
                  data=df_dcharge,
                  color='#0057a8',
                  )
plt.legend(labels=['Stream Gauge Flow'])

# appplies labels and tick formatting
plt.xticks(rotation=45)
plt.title('North Cunningham Creek Stream Gauge Flow Sep 1st through Aug 31st for 2014-2015',
          weight='bold',
          y=1.03).set_fontsize('19')
plt.xlabel('Month')
plt.ylabel('Discharge (m\u00b3/s)')

# sets the major ticks for the first day of each month, must import matplotlib.dates function 
ax.xaxis.set_major_locator(mdates.MonthLocator(bymonthday=1))

# formats the major ticks to month only, must import matplotlib.dates function
ax.xaxis.set_major_formatter(mdates.DateFormatter('%B'))

plt.show()
```

**1d. For the gauges with missing data, the missing days occur between about October and March. Why might streamflow values be missing during those months?** 

North Cunningham Creek is missing stream gauge flow data starting on October and through March. This is most likely due to below freezing temperatures that caused the stream to freeze over the winter and not warm back up until April where the discharge can be recorded.
